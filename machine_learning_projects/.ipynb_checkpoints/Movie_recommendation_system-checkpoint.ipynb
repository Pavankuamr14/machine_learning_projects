```python
movies=movies.merge(credits,on='title')
```

```python
import ast
```

```python
var=movies.iloc[1]['genres']
```

```python
def genres_data(data):
    lst=[]
    for i in ast.literal_eval(var):
        lst.append(i['name'])
    return lst
```

```python
movies['genres']=movies['genres'].apply(genres_data)
```

```python
movies
```

```python
movies.iloc[1]['keywords']
```

```python
def keywords(data):
    lst=[]
    for i in ast.literal_eval(data):
        lst.append(i['name'])
    return lst
        
```

```python
movies['keywords']=movies['keywords'].apply(keywords)
```

```python
movies
```

```python
def space_remover(data):
    lst=[]
    for i in data:
        lst.append(i.replace(" ",""))
    return lst
```

```python
movies['cast'] = movies['cast'].apply(space_remover)
movies['crew'] = movies['crew'].apply(space_remover)
movies['genres']=movies['genres'].apply(space_remover)
movies['keywords']=movies['keywords'].apply(space_remover)
```

```python
def actors(data):
    lst=[]
    for i in ast.literal_eval(data):
        if i['order'] < 5:
            lst.append(i['order'])
    return lst
```
